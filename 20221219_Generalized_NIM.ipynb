```python
def winable(x,y):
    if x<0 or y<0:
        return True
    if x==0:
        return False
    return not(winable(y,x-1) and winable(x-2,y))
```

```python
for x in range(40):
    print(x)
    for y in range(x):
        print(winable(x,y))
    print()
```
